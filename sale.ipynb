```python
import pandas as pd
df=pd.read_csv("sales_dirty_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Order_Date"]=pd.to_datetime(df["Order_Date"],dayfirst=True, errors='coerce')
```

```python
df["Order_Date"].head()
```

```python
df["Customer_Name"]=df["Customer_Name"].fillna("Unknown")
```

```python
df["Customer_Name"].head()
```

```python
df["Payment_Mode"]=df["Payment_Mode"].str.lower()
```

```python
df["Payment_Mode"]=df["Payment_Mode"].replace({"card":"Card","cash":"Cash","upi":"UPI"})
```

```python
df["Payment_Mode"]=df["Payment_Mode"].fillna("Card")
```

```python
df["Product"]=df["Product"].replace({"Mobile Phone":"Mobile"})
```

```python
df["Quantity"]=df["Quantity"].fillna(1)
```

```python
df=df[df["Quantity"]>0].copy()
```

```python
df["Total_Amount"]=df["Quantity"]*df["Price"]
```

```python
df["City"]=df["City"].fillna("Unknown")
```

```python
df=df.drop_duplicates()
```
